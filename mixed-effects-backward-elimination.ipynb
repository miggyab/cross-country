```r
library(lmerTest)
```

```r
df = read.csv('../input/country-metrics/country_pairs.csv')

df$diff_sample = scale(df$diff_sample)
df$diff_gdp_bill = scale(df$diff_gdp_bill)
df$diff_happiness = scale(df$diff_happiness)
df$diff_population = scale(df$diff_population)
df$diff_per_capita = scale(df$diff_per_capita)
```

```r
model = lmerTest::lmer(auc_score ~ diff_sample + diff_happiness + diff_gdp_bill + diff_population + diff_per_capita + (1|train_country), data = df)
```

```r
step_res = lmerTest::step(model)
step_res
```

```r
final = get_model(step_res)
anova(final)
```
