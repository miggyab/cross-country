```r
library(lmerTest)
```

```r
df = read.csv('../input/country-metrics/country_pairs.csv')

df$diff_sample = scale(df$diff_sample)
df$diff_gdp_bill = scale(df$diff_gdp_bill)
df$diff_happiness = scale(df$diff_happiness)
df$diff_population = scale(df$diff_population)
df$diff_per_capita = scale(df$diff_per_capita)

df$diff_pdi = scale(df$diff_pdi)
df$diff_idv = scale(df$diff_idv)
df$diff_mas = scale(df$diff_mas)
df$diff_uai = scale(df$diff_uai)
df$diff_ltowvs = scale(df$diff_ltowvs)
df$diff_ivr = scale(df$diff_ivr)
```

```r
model = lmerTest::lmer(distance ~ diff_sample + diff_happiness + diff_gdp_bill + diff_population + diff_per_capita + (1|train_country), data = df)
```

```r
step_res = lmerTest::step(model)
step_res
```

```r
final = get_model(step_res)
anova(final)
```

```r
hof_mod = lmerTest::lmer(distance ~ diff_pdi + diff_idv + diff_mas + diff_uai + diff_ltowvs + diff_ivr + (1|train_country), data=df)
```

```r
step_res = lmerTest::step(hof_mod)
step_res
```

```r
final = get_model(step_res)
anova(final)
```
